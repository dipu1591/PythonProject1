```python
dataset1= dataset.drop_duplicates()                                             # Removes duplicate rows from dataset and stores the result in `dataset1`.

dataset1.shape                                                                  # Retrieves and prints the dimensions of `dataset1`, showing the number of rows and columns after removing duplicates.
```

```python
dataset1.info()                                                                 # Displays a concise summary of `dataset1`, including index dtype and columns, non-null values, and memory usage.
```

```python
dataset1['OCCURRED_ON_DATE']= pd.to_datetime(dataset1['OCCURRED_ON_DATE'])      # Converts the 'OCCURRED_ON_DATE' column in `dataset1` to datetime objects for proper date-time handling.

dataset1.info()                                                                 # Displays a summary of `dataset1` to verify changes, such as the data type conversion of 'OCCURRED_ON_DATE'.
```

```python
dataset1.head()                                                                 # Displays the first five rows of `dataset1` to provide a quick overview of its data structure and content.
```

```python
dataset1['Year']= dataset1['OCCURRED_ON_DATE'].dt.year                                                  # Extracts Year from the 'OCCURRED_ON_DATE' column and creates new column named 'Year'

dataset1.insert(dataset1.columns.get_loc('OCCURRED_ON_DATE') + 1, 'Year', dataset1.pop('Year'))         # Reorders the DataFrame by inserting the newly created 'Year' column right after the 'OCCURRED_ON_DATE' column

dataset1['Month']= dataset1['OCCURRED_ON_DATE'].dt.month                                                # Extracts Month number from the 'OCCURRED_ON_DATE' column and creates new column named 'Month'

dataset1.insert(dataset1.columns.get_loc('Year') + 1, 'Month', dataset1.pop('Month'))                   # Reorders the DataFrame by inserting the newly created 'Month' column right after the 'Year' column

dataset1['Week']= dataset1['OCCURRED_ON_DATE'].dt.week                                                  # Extracts Week number from the 'OCCURRED_ON_DATE' column and creates new column named 'Week'

dataset1.insert(dataset1.columns.get_loc('Month') + 1, 'Week', dataset1.pop('Week'))                    # Reorders the DataFrame by inserting the newly created 'Week' column right after the 'Month' column

dataset1['Day']= dataset1['OCCURRED_ON_DATE'].dt.day                                                    # Extracts Date from the 'OCCURRED_ON_DATE' column and creates new column named 'Day'

dataset1.insert(dataset1.columns.get_loc('Week') + 1, 'Day', dataset1.pop('Day'))                       # Reorders the DataFrame by inserting the newly created 'Day' column right after the 'Week' column

dataset1['Time']= dataset1['OCCURRED_ON_DATE'].dt.time                                                  # Extracts Time from the 'OCCURRED_ON_DATE' column and creates new column named 'Time'

dataset1.insert(dataset1.columns.get_loc('Day') + 1, 'Time', dataset1.pop('Time'))                      # Reorders the DataFrame by inserting the newly created 'Time' column right after the 'Day' column

dataset1['Hour']= dataset1['OCCURRED_ON_DATE'].dt.hour                                                  # Extracts Hour from the 'OCCURRED_ON_DATE' column and creates new column named 'Hour'

dataset1.insert(dataset1.columns.get_loc('Time') + 1, 'Hour', dataset1.pop('Hour'))                     # Reorders the DataFrame by inserting the newly created 'Hour' column right after the 'Day' column

month_names= {                                                                                          # Defines a dictionary mapping month numbers to month names
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

dataset1['Month name']= dataset1['Month'].map(month_names)                                              # Maps the 'Month' column to their respective month names using the dictionary created and adds a new column 'Month name'.

dataset1.head()                                                                                         # Displays the first five rows of `dataset1` to show the updated structure with the newly added and reordered columns.
```

# ***1. What are the most common crimes reported in terms of offense group?***



```python
ocg_count= dataset1['OFFENSE_CODE_GROUP'].value_counts()                                                                              # Counts occurrences of each unique value in the 'OFFENSE_CODE_GROUP' column of `dataset1` and stores the result.

ocg_count_percent= (ocg_count / len(dataset1))*100                                                                                    # Calculates the percentage of each offense code group relative to the total number of rows in the `dataset1`.

ocg_count_df= pd.DataFrame({'Category': ocg_count.index, 'Count': ocg_count.values, 'Percentage': ocg_count_percent.values})          # Creates a DataFrame from the counts and percentages, labeling columns as 'Category', 'Count', and 'Percentage'.

ocg_count_df['Percentage']= ocg_count_df['Percentage'].round(2).astype(str) + '%'                                                     # Rounds the 'Percentage' column to two decimal places, converts it to a string, and appends a percentage sign.

pd.set_option('display.max_rows', None)                                                                                               # Configures pandas to display all rows of the DataFrame without truncation.

ocg_count_df                                                                                                                          # Displays the ocg_count_df DataFrame, showing the offense code group categories, their counts, and percentages of the total incidents.
```

# ***2. How has the number of crimes changed over the years?***

```python
incident_counts= dataset1.groupby('Year').count()['INCIDENT_NUMBER']                                # Groups the data by 'Year' and counts the number of incidents in each year, storing the result in incident_counts.

assign_colors= plt.cm.cividis(np.linspace(0, 1, len(incident_counts)))                              # Generates a color gradient using the Cividis colormap, with the number of colors equal to the number of years in incident_counts.

plt.bar(incident_counts.index.astype(str), incident_counts, color= assign_colors)                   # Creates a bar chart with years on the x-axis and the number of crimes on the y-axis, coloring each bar with the previously generated color gradient.

plt.title("Number of crimes in terms of year", pad= 25)                                             # Sets the title of the plot and adds padding to avoid overlap with the plot elements.

plt.xlabel("Year", labelpad= 25)                                                                    # Labels the x-axis as "Year" and adds padding to avoid overlap with the plot elements.

plt.ylabel("Number of crimes", labelpad= 30)                                                        # Labels the y-axis as "Number of crimes" and adds padding to avoid overlap with the plot elements.

plt.grid(axis= 'y', linestyle= '--', linewidth= 0.5, color= 'gray')                                 # Adds a grid to the y-axis with a specified style, linewidth, and color to help in visually assessing the number of crimes.
```

# ***3. Do more crimes occur on certain dates of the month?***

```python
month_to_day= dataset1.groupby(['Month name','Day']).count()['INCIDENT_NUMBER'].unstack()                 # Groups dataset1 by 'Month name' and 'Day', counts the number of incidents for each day of each month, and then reshapes the result into a pivot table format where each row represents a month, and each column represents a day.

pd.set_option('display.max_columns', None)                                                                # Configures pandas to display an unlimited number of columns when printing DataFrames, ensuring that the entire month_to_day pivot table is visible without truncation.

month_to_day                                                                                              # Displays the month_to_day DataFrame
```

```python
month_order= ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']           # Defines the correct chronological order for the months.

month_to_day_row_reordered= month_to_day.reindex(month_order)                                                                                     # Reorders the rows of the month_to_day DataFrame to match the chronological order specified in month_order.

plt.figure(figsize= (22, 6))                                                                                                                      # Creates a new figure with a specified size to ensure the heatmap is large enough to be easily readable.

sns.heatmap(month_to_day_row_reordered, annot= True, fmt='.0f')                                                                                   # Generates a heatmap from the reordered DataFrame, annotating each cell with the integer number of incidents (no decimal points) to show the distribution of incidents across different days and months.

plt.title('Heatmap of Incidents by Month and Day', pad= 25)                                                                                       # Sets the title of the heatmap and adds padding to avoid overlap with the plot elements.

plt.xlabel('Date', labelpad= 30)                                                                                                                  # Labels the x-axis as 'Date' and adds padding to avoid overlap with the plot elements.

plt.ylabel('Month', labelpad= 32)                                                                                                                 # Labels the y-axis as 'Month' and adds padding to avoid overlap with the plot elements.
```

# ***4. Have there been more crimes in any month over the past years?***

```python
month_to_year= dataset1.groupby(['Year','Month name']).count()['INCIDENT_NUMBER'].unstack()                                                       # Groups the data by 'Year' and 'Month name', counts incidents for each month-year combination, and reshapes the data into a pivot table where rows are years and columns are months.

month_order= ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']           # Defines the chronological order of months for later reordering of columns.

month_to_year_column_reordered= month_to_year.reindex(columns= month_order)                                                                       # Reorders the columns of the pivot table according to the chronological order of months specified in month_order

plt.figure(figsize= (17, 3))                                                                                                                      # Initializes a new figure for plotting with a specified figure size.

sns.heatmap(month_to_year_column_reordered, annot= True, fmt= '.0f')                                                                              # Creates a heatmap using the reordered pivot table, with annotations to display the count of incidents in each cell, formatted as integers.

plt.title('Heatmap of Incidents by Month and Year', pad= 25)                                                                                      # Sets the title of the heatmap and adds padding to avoid overlap with the plot elements.

plt.xlabel('Month', labelpad= 30)                                                                                                                 # Labels the x-axis as 'Month' and adds padding to avoid overlap with the plot elements.

plt.ylabel('Year', labelpad= 32)                                                                                                                  # Labels the y-axis as 'Year' and adds padding to avoid overlap with the plot elements.
```

# ***5. Does crime tend to increase at certain times of the day?***

```python
week_to_hour= dataset1.groupby(['DAY_OF_WEEK','Hour']).count()['INCIDENT_NUMBER'].unstack()               # Groups 'dataset1' by 'DAY_OF_WEEK' and 'Hour', counts the number of incidents for each hour of each day, and then reshapes the result into a pivot table format where each row represents a day of the week, and each column represents an hour of the day.

pd.set_option('display.max_columns', None)                                                                # Configures pandas to display an unlimited number of columns when printing DataFrames, ensuring that the entire week_to_hour pivot table is visible without truncation.

week_to_hour                                                                                              # Displays the week_to_hour DataFrame
```

```python
day_order= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']         # Defines the order of days from Monday to Sunday for later reordering of rows.

week_to_hour_row_reordered= week_to_hour.reindex(day_order)                                       # Reorders the rows of the pivot table according to the specified order of days in day_order.

plt.figure(figsize= (22, 6))                                                                      # Initializes a new figure for plotting with a specified figure size.

sns.heatmap(week_to_hour_row_reordered, annot= True, fmt= '.0f')                                  # Creates a heatmap using the reordered pivot table, with annotations to display the count of incidents in each cell, formatted as integers.

plt.title('Heatmap of Incidents by Day and Hour', pad= 25)                                        # Sets the title of the heatmap and adds padding to avoid overlap with the plot elements.

plt.xlabel('Hour', labelpad= 30)                                                                  # Labels the x-axis as 'Hour' and adds padding to avoid overlap with the plot elements.

plt.ylabel('Day', labelpad= 32)                                                                   # Labels the y-axis as 'Day' and adds padding to avoid overlap with the plot elements.
```

# ***6. In which months were the number of crimes above average?***

```python
avg_num_of_crime= month_to_year_column_reordered.copy()                                               # Creates a copy of the 'month_to_year_column_reordered' DataFrame to preserve the original data, naming avg_num_of_crime.

avg_num_of_crime['Average']= avg_num_of_crime.loc[:, 'January':'December'].mean(axis=1)               # Calculates the mean number of crimes for each year across all months (from January to December) and adds these values as a new column called 'Average' in the avg_num_of_crime DataFrame.

avg_num_of_crime                                                                                      # Displays the modified avg_num_of_crime DataFrame, now including the average number of crimes per year.
```

```python
higher_average= avg_num_of_crime.loc[:, 'January':'December'].lt(avg_num_of_crime['Average'], axis= 0)                                    # Compares each month's crime count against the yearly average within avg_num_of_crime. It produces a boolean DataFrame (higher_average) where True indicates a month's crime count is less than the average, and False indicates it is equal to or higher.

plt.figure(figsize= (15, 3))                                                                                                              # Initializes a new figure for plotting with a specified figure size.

sns.heatmap(higher_average, cmap= ['red', 'black'], annot= avg_num_of_crime.loc[:, 'January':'December'], fmt= 'd', cbar= False)          # Creates a heatmap using the boolean DataFrame higher_average to color-code months. Months with crime counts lower than the yearly average are colored black (True), and those equal to or higher are colored red (False). Actual crime counts are annotated on the heatmap, with no color bar displayed (cbar=False).

plt.title('Comparison of monthly crime counts with average  (Black: Lower, Red: Higher)', pad= 25)                                        # Sets the title of the heatmap and adds padding to avoid overlap with the plot elements.

plt.xlabel('Month', labelpad= 30)                                                                                                         # Labels the x-axis as 'Month' and adds padding to avoid overlap with the plot elements.

plt.ylabel('Year', labelpad= 32)                                                                                                          # Labels the y-axis as 'Year' and adds padding to avoid overlap with the plot elements.
```

# ***7. Are certain times of the day more prone to specific types of crimes?***

```python
crime_by_time= dataset1.groupby(['Hour', 'OFFENSE_CODE_GROUP']).size().reset_index(name= 'COUNT')                             # Groups 'dataset1' by 'Hour' and 'OFFENSE_CODE_GROUP', calculates the size of each group, and resets the index to convert the Series into a DataFrame named crime_by_time, with a new column 'COUNT' representing the size of each group.

crime_by_time_pivot= crime_by_time.pivot(index= 'OFFENSE_CODE_GROUP', columns= 'Hour', values= 'COUNT').fillna(0)             # Transforms crime_by_time into a pivot table (crime_by_time_pivot) where each row represents an offense code group, each column represents an hour of the day, and cell values represent the count of crimes. Missing values are filled with 0.

crime_by_time_pivot                                                                                                           # Displays the pivot table created, showing the distribution of crime counts across different hours for each offense code group.
```

```python
plt.figure(figsize= (16, 10))                                                   # Initializes a new figure for the heatmap with a specified size

sns.heatmap(crime_by_time_pivot)                                                # Generates a heatmap using the crime_by_time_pivot DataFrame, which contains the counts of offenses by hour for each offense type.

plt.title('Heatmap of offense types by time', pad= 25)                          # Sets the title of the heatmap and adds padding to avoid overlap with the plot elements.

plt.xlabel('Time', labelpad= 30)                                                # Labels the x-axis as "Time" and adds padding to avoid overlap with the plot elements.

plt.ylabel('Offense Types', labelpad= 32)                                       # Labels the y-axis as "Offense Types" and adds padding to avoid overlap with the plot elements.
```

# ***8. Which districts have the highest number of crimes?***

```python
district_counts= dataset1.groupby('DISTRICT')['INCIDENT_NUMBER'].count().sort_values(ascending= True).reset_index()                 # Groups dataset1 by 'DISTRICT', counts the number of incidents in each district, sorts these counts in ascending order, and resets the index to convert the Series to a DataFrame.

num_colors= district_counts.shape[0]                                                                                                # Determines the number of unique districts by getting the number of rows in district_counts, used to generate a color palette with a distinct color for each district.

color_palette= sns.cubehelix_palette(start= .5, rot= -.5, as_cmap= False, n_colors= num_colors)                                     # Generates a sequential color palette using the cubehelix system, with a specific starting point and rotation, and assigns a unique color to each district.

sns.barplot(x= 'INCIDENT_NUMBER', y= 'DISTRICT', palette= color_palette, data= district_counts)                                     # Creates a horizontal bar plot with the number of crimes on the x-axis and districts on the y-axis, applying the generated color palette to distinguish between districts.

plt.xlabel("Number of crime", labelpad= 30)                                                                                         # Labels the x-axis as "Number of crime" and adds padding to avoid overlap with the plot elements.

plt.ylabel("District", labelpad= 22)                                                                                                # Labels the y-axis as "District" and adds padding to avoid overlap with the plot elements.

plt.title("Number of crimes in different districts", pad= 25)                                                                       # Sets the title of the plot and adds padding to avoid overlap with the plot elements.

plt.grid(axis= 'x', linestyle= '--', linewidth= 0.5, color= 'gray')                                                                 # Adds a horizontal grid to the plot to facilitate the comparison of bar lengths, using a dashed line style and gray color for subtlety.
```
